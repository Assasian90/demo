```python
import numpy as np
import pandas as pd
data={
    "name":["Alice","Bob","Charlie","David","Eve","Alice"],
    "Age":[25,30,35,np.nan,29,25],
    "Department":["HR","IT","Finance","IT","HR","HR"],
    "Salary":[50000,60000,70000,62000,np.nan,50000]
}
df=pd.DataFrame(data)
df
```

```python
df_dup=df[df.duplicated(keep="last")]
df_dup
```

```python
df.drop_duplicates()
```

```python
df["prmoted salary"]=df["Salary"]*10
df
```

```python
df['prmoted salary']=df["prmoted salary"].apply(lambda x:x/10 if x>650000 else x)
df
```

```python
df['Age'].fillna(df['Age'].median())
df
```

```python
df["Age"]=df["Age"].apply(lambda x:x*2)
df
```

```python
df["Age"]=df["Age"].apply(lambda x:x/4)
df
```
